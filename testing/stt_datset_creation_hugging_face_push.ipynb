```python
import json
import webvtt

# Replace 'subtitles.vtt' with the actual path to your VTT file
vtt_file = "الحلقة الخامسة： توقف الحرب.. واستمرارها - بودكاست مع تميم [1Zaxz19PP-I].ar.vtt"
json_file = "subtitles.json"

# Helper function to convert time string to seconds
def time_to_seconds(time_str):
    parts = time_str.split(":")
    if len(parts) == 3:  # Format is HH:MM:SS.mmm
        hours, minutes, seconds = parts
    elif len(parts) == 2:  # Format is MM:SS.mmm
        hours = 0
        minutes, seconds = parts
    else:
        raise ValueError(f"Unexpected time format: {time_str}")
    return int(hours) * 3600 + int(minutes) * 60 + float(seconds)

# Parse the VTT file
subtitles = []
for caption in webvtt.read(vtt_file):
    subtitles.append({
        "text": caption.text.strip(),
        "start": time_to_seconds(caption.start),
        "end": time_to_seconds(caption.end)
    })

# Save to JSON
with open(json_file, "w", encoding="utf-8") as f:
    json.dump(subtitles, f, ensure_ascii=False, indent=4)

print(f"Subtitles saved to {json_file}")
```

```python
import json
import os
import pandas as pd
import boto3
from botocore.exceptions import NoCredentialsError
import subprocess
from io import BytesIO

# AWS S3 configuration
s3_bucket_name = 'stt-data-large'
s3_audio_folder = 'stt-dataset-v31/'  # Optional folder path inside the bucket

# AWS S3 client
s3_client = boto3.client('s3')

# Verify JSON file exists and is non-empty
json_file = 'subtitles.json'
if not os.path.exists(json_file):
    raise FileNotFoundError(f"{json_file} does not exist.")
if os.stat(json_file).st_size == 0:
    raise ValueError(f"{json_file} is empty. Please provide valid JSON data.")

# Load the transcription data from the JSON file
with open(json_file, 'r', encoding='utf-8') as f:
    try:
        transcription_data = json.load(f)
    except json.JSONDecodeError as e:
        raise ValueError(f"Error decoding JSON: {e}")
        
# Load the audio file (assuming it's a .webm file, which ffmpeg supports)
audio_file = 'الحلقة الخامسة： توقف الحرب.. واستمرارها - بودكاست مع تميم [1Zaxz19PP-I].webm'
if not os.path.exists(audio_file):
    raise FileNotFoundError(f"{audio_file} does not exist.")

# List to store the data
data = []

# Process each transcription entry and extract audio segments
for entry in transcription_data:
    text = entry['text']
    start_time = entry['start']  # Start time in seconds
    end_time = entry['end']      # End time in seconds

    # Construct the filename based on the start and end times
    audio_segment_filename = f"stt-dataset-v31_{int(start_time)}_{int(end_time)}.wav"

    # Use ffmpeg to extract the segment from the large audio file in memory
    command = [
        'ffmpeg', '-i', audio_file,  # Input file
        '-ss', str(start_time),      # Start time (in seconds)
        '-to', str(end_time),        # End time (in seconds)
        '-f', 'wav',                 # Output format
        '-'                          # Output to stdout (in memory)
    ]
    
    # Execute the command and capture the output
    try:
        process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        audio_data, err = process.communicate()
        
        if process.returncode != 0:
            raise ValueError(f"Error during ffmpeg extraction: {err.decode('utf-8')}")

        # Create BytesIO object for the audio data
        audio_bytes_io = BytesIO(audio_data)

        # Upload the audio chunk to S3 directly
        s3_client.upload_fileobj(audio_bytes_io, s3_bucket_name, s3_audio_folder + audio_segment_filename)
        
        # Append transcription and S3 URL to the data list
        audio_s3_url = f"s3://{s3_bucket_name}/{s3_audio_folder}{audio_segment_filename}"
        data.append({
            "text": text,
            "audio_file": audio_s3_url  # This stores the S3 URL of the audio file
        })

    except subprocess.CalledProcessError as e:
        raise ValueError(f"Error extracting audio segment: {e}")

# Create a DataFrame for the final dataset (this can be saved as CSV or any other format)
df = pd.DataFrame(data)

# Define the path for the .csv file (if needed for later use)
stt_dataset_path = 'stt_dataset_v31.csv'
df.to_csv(stt_dataset_path, index=False)

# Save the transcription text-only in a .txt file (with one line per entry)
text_only_path = 'transcription_text_only_v31.txt'
with open(text_only_path, 'w', encoding='utf-8') as f:
    for entry in data:
        f.write(f"{entry['text']}\n")

# Print out where the files have been saved
print(f"Processed audio chunks uploaded to S3 bucket '{s3_bucket_name}'")
print(f"Text-only transcriptions saved to {text_only_path}")
```

```python
import pandas as pd
import os
import re
import unicodedata
from datasets import Dataset, DatasetDict, concatenate_datasets
from datasets import Audio
from num2words import num2words

# Define the path to your CSV file and audio directory
csv_path = "/content/drive/MyDrive/stt_dataset_v30.csv"
audio_dir = "/content/drive/MyDrive/stt-dataset-v30"

# Load your CSV file with transcriptions
df = pd.read_csv(csv_path)

# Clean up column names to remove leading/trailing spaces
# df.columns = df.columns.str.strip()

# List all the .wav files in the directory
wav_files = sorted([f for f in os.listdir(audio_dir) if f.endswith('.wav')])

# Function to normalize Arabic transcription
def normalize_arabic(text):
    # Remove unwanted symbols (non-arabic characters)
    text = re.sub(r"[^\w\s\u0621-\u064A\u0660-\u0669]", "", text)

    # Normalize Arabic script (remove diacritics, etc.)
    text = unicodedata.normalize("NFKD", text)  # Normalize Arabic text to a consistent form

    # Optional: Convert Arabic numbers to words
    text = re.sub(r"\d+", lambda x: num2words(x.group(), lang='ar'), text)

    return text.strip()

# Apply normalization to the transcriptions
df['text'] = df['text'].apply(normalize_arabic)

# Initialize a list to store the processed data
data = []

# Iterate over the rows of the dataframe
for idx, row in df.iterrows():
    # Assuming 'audio_path' column contains the filename without the directory path
    audio_filename = row["audio_file"].strip()

    # Construct the full path of the audio file
    audio_path = os.path.join(audio_dir, audio_filename)

    # Check if the audio file exists in the specified directory
    if os.path.exists(audio_path):
        # Append the audio path and transcription to the data list
        data.append({
            "audio": audio_path,  # Full path to the audio file
            "text": row["text"]  # Corresponding transcription
        })
    else:
        print(f"Warning: Audio file '{audio_filename}' not found at path '{audio_path}'")

# Check if data is populated
if len(data) == 0:
    print("Error: No valid data found. Please check the audio paths.")
else:
    print(f"Loaded {len(data)} audio files and transcriptions.")

# Create a Hugging Face Dataset from the list of dictionaries
dataset = Dataset.from_list(data)

# Load audio feature (this will automatically handle the .wav files)
dataset = dataset.cast_column("audio", Audio(sampling_rate=16000))  # Resample to 16kHz if necessary

# Split the dataset into 80% training, 10% testing, and 10% validation
train_test_split = dataset.train_test_split(test_size=0.2, seed=42)  # 20% for test set

# Further split the test set into validation and test (50% of test set -> 10% validation, 10% test)
test_val_split = train_test_split["test"].train_test_split(test_size=0.5, seed=42)  # 50% of test set -> 10% validation, 10% test

# Combine the splits back into a DatasetDict
dataset_dict = DatasetDict({
    "train": train_test_split["train"],
    "test": test_val_split["test"],
    "validation": test_val_split["train"]
})


dataset_dict.push_to_hub("crtvai/jordan-dataset-v30")

# # print("Dataset successfully prepared and pushed to Hugging Face Hub.")
print(dataset_dict)
```

```python
import pandas as pd
import os
from datasets import Dataset, DatasetDict, Audio
from datasets import load_dataset
from datasets import load_dataset, concatenate_datasets
# Load your CSV file with transcriptions
csv_path = "/content/drive/MyDrive/updated_stt_dataset_with_arabic-five.csv"
df = pd.read_csv(csv_path)

# Clean up column names to remove leading/trailing spaces
df.columns = df.columns.str.strip()

# List all the .wav files in the directory
audio_dir = "/content/drive/MyDrive/stt-dataset-version-five"
wav_files = sorted([f for f in os.listdir(audio_dir) if f.endswith('.wav')])

# Map CSV filenames (e.g., 0.wav, 1.wav, ...) to actual .wav filenames
audio_files = {}
for idx, row in df.iterrows():
    # Assuming the CSV filenames are sequential, match by index
    simple_name = row["audio_path"]

    if idx < len(wav_files):  # Ensure we don't go out of bounds
        # Get the actual audio file name from the directory
        actual_audio_path = os.path.join(audio_dir, wav_files[idx])
        audio_files[actual_audio_path] = row["transcription"]
    else:
        print(f"Warning: No audio file found for index {idx}.")

# Initialize a list to store the processed data
data = []

# Iterate over the audio filenames and transcriptions
for audio_path, transcription in audio_files.items():
    # Check if the audio file exists and append to the data list
    if os.path.exists(audio_path):
        data.append({
            "audio": audio_path,  # Use the full audio path
            "transcription": transcription
        })
    else:
        print(f"Warning: Audio file '{audio_path}' not found in {audio_dir}")

# Check if data is populated
if len(data) == 0:
    print("Error: No valid data found. Please check the audio paths.")
else:
    print(f"Loaded {len(data)} audio files and transcriptions.")

# Create a Hugging Face Dataset from the list of dictionaries
dataset = Dataset.from_list(data)

# Load audio feature (this will automatically handle the .wav files)
dataset = dataset.cast_column("audio", Audio())

# Split the dataset into 80% training, 10% testing, and 10% validation
train_test_split = dataset.train_test_split(test_size=0.2, seed=42)  # 20% for test set

# Now split the 20% test data into 50% test and 50% validation, which is 10% each of the total dataset
test_val_split = train_test_split["test"].train_test_split(test_size=0.5, seed=42)  # 50% of test set -> 10% validation, 10% test

# Combine the splits back into a DatasetDict
dataset_dict = DatasetDict({
    "train": train_test_split["train"],
    "test": test_val_split["test"],
    "validation": test_val_split["train"]
})

# Load existing dataset from the Hugging Face Hub
from datasets import DatasetDict
existing_dataset = load_dataset("Fahim18/stt-dataset-two")

# Merge the new splits with the existing ones
existing_train = existing_dataset["train"]
existing_test = existing_dataset["test"]
existing_validation = existing_dataset["validation"]

# Merge the new dataset with the existing one
new_train = concatenate_datasets([existing_train, dataset_dict["train"]])
new_test = concatenate_datasets([existing_test, dataset_dict["test"]])
new_validation = concatenate_datasets([existing_validation, dataset_dict["validation"]])

# Combine all splits back into a DatasetDict
final_dataset_dict = DatasetDict({
    "train": new_train,
    "test": new_test,
    "validation": new_validation
})

# Push the combined dataset back to Hugging Face
final_dataset_dict.push_to_hub("Fahim18/stt-dataset-two")
```

```python
# Resample audio function
def resample_audio(example, target_sampling_rate):
    audio_key = 'audio'  # Explicitly set the audio column name
    audio_data = example[audio_key]

    # Ensure that 'audio' data contains necessary fields
    if isinstance(audio_data, dict) and "array" in audio_data and "sampling_rate" in audio_data:
        # If the sampling rate is missing or None, skip resampling for this example
        if audio_data["sampling_rate"] is None:
            print(f"Warning: Missing sampling_rate for example with audio_data: {audio_data}")
            return example

        # If resampling is needed, proceed with resampling
        if audio_data["sampling_rate"] != target_sampling_rate:
            resampled_audio = librosa.resample(
                audio_data["array"],
                orig_sr=audio_data["sampling_rate"],
                target_sr=target_sampling_rate
            )
            example[audio_key] = {"array": resampled_audio, "sampling_rate": target_sampling_rate}
    else:
        print(f"Warning: Skipping resampling for example with missing audio_data or sampling_rate: {audio_data}")
    return example

# Set target_sampling_rate (if it's None, default to 16000)
target_sampling_rate = existing_dataset["train"].features['audio'].sampling_rate if existing_dataset["train"].features['audio'].sampling_rate is not None else 16000
print(f"Using target sampling rate: {target_sampling_rate} Hz")

# Apply resampling function to each example in the dataset
dataset_dict = dataset_dict.map(lambda x: resample_audio(x, target_sampling_rate))

def cast_audio_to_target_rate(example, target_sampling_rate):
    audio_key = 'audio'
    audio_data = example[audio_key]

    # Resample if necessary
    if audio_data["sampling_rate"] != target_sampling_rate:
        resampled_audio = librosa.resample(
            audio_data["array"],
            orig_sr=audio_data["sampling_rate"],
            target_sr=target_sampling_rate
        )
        example[audio_key] = {"array": resampled_audio, "sampling_rate": target_sampling_rate}

    return example

# Ensure both datasets have the same sampling rate before concatenation
target_sampling_rate = 16000  # or your desired rate

# Apply to both datasets
dataset_dict = dataset_dict.map(lambda x: cast_audio_to_target_rate(x, target_sampling_rate))
existing_dataset = existing_dataset.map(lambda x: cast_audio_to_target_rate(x, target_sampling_rate))

```

```python
import yt_dlp
import json
import os
import requests
import sqlite3
import pandas as pd
from pydub import AudioSegment
from concurrent.futures import ThreadPoolExecutor

# Define global variables
OUTPUT_DIR = "/content/drive/MyDrive/stt-dataset-v31"
DB_PATH = "/content/drive/MyDrive/stt_dataset_v31.db"
CHUNK_SIZE = 5

# Ensure the output directory exists
os.makedirs(OUTPUT_DIR, exist_ok=True)

# SQLite setup
def setup_database():
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()

    # Create tables if they don't exist
    cursor.execute('''CREATE TABLE IF NOT EXISTS videos (
                        id INTEGER PRIMARY KEY,
                        url TEXT,
                        subtitle_path TEXT,
                        audio_path TEXT,
                        processed INTEGER DEFAULT 0)''')

    cursor.execute('''CREATE TABLE IF NOT EXISTS transcriptions (
                        id INTEGER PRIMARY KEY,
                        video_id INTEGER,
                        start_time REAL,
                        end_time REAL,
                        text TEXT,
                        audio_chunk_path TEXT,
                        FOREIGN KEY (video_id) REFERENCES videos(id))''')

    conn.commit()
    conn.close()

# Insert video URLs into the database
def insert_video_urls(video_urls):
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    for url in video_urls:
        cursor.execute("INSERT INTO videos (url) VALUES (?)", (url,))
    conn.commit()
    conn.close()

# Fetch unprocessed video URLs from the database
def fetch_unprocessed_videos():
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    cursor.execute("SELECT id, url FROM videos WHERE processed = 0")
    videos = cursor.fetchall()
    conn.close()
    return videos

# Update video processing status
def mark_video_as_processed(video_id):
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()
    cursor.execute("UPDATE videos SET processed = 1 WHERE id = ?", (video_id,))
    conn.commit()
    conn.close()

# Function to process a single video
def process_video(video_id, video_url):
    ydl_opts = {
        'format': 'bestaudio/best',
        'outtmpl': os.path.join(OUTPUT_DIR, f"audio_{video_id}.%(ext)s"),
        'writesubtitles': True,
        'writeautomaticsub': True,
        'skip_download': False,
        'verbose': True,
    }

    try:
        # Download video and extract subtitles
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info_dict = ydl.extract_info(video_url, download=True)

            # Paths for subtitles and audio
            subtitle_path = None
            audio_path = ydl_opts['outtmpl'].replace('%(ext)s', info_dict['ext'])

            # Find Arabic subtitles
            subtitles = info_dict.get('subtitles', {})
            automatic_captions = info_dict.get('automatic_captions', {})

            if 'ar' in subtitles:
                subtitle_path = subtitles['ar'][0]['url']
            elif 'ar' in automatic_captions:
                subtitle_path = automatic_captions['ar'][0]['url']

            # Store video metadata in the database
            conn = sqlite3.connect(DB_PATH)
            cursor = conn.cursor()
            cursor.execute("UPDATE videos SET subtitle_path = ?, audio_path = ? WHERE id = ?", 
                           (subtitle_path, audio_path, video_id))
            conn.commit()
            conn.close()

            # Process subtitles if available
            if subtitle_path:
                response = requests.get(subtitle_path)
                response.raise_for_status()
                subtitle_data = response.json()

                transcription = []
                last_start_time = None

                # Parse subtitles
                for event in subtitle_data.get('events', []):
                    start = event.get('tStartMs', 0) / 1000
                    text_segments = event.get('segs', [])
                    text = ''.join(seg.get('utf8', '') for seg in text_segments)

                    if text.strip():
                        if last_start_time is not None:
                            transcription[-1]['end'] = start

                        transcription.append({
                            'start': start,
                            'text': text.strip(),
                        })
                        last_start_time = start

                if transcription:
                    transcription[-1]['end'] = last_start_time

                # Save transcription and create audio chunks
                save_transcription_and_chunks(video_id, audio_path, transcription)

            mark_video_as_processed(video_id)

    except Exception as e:
        print(f"An error occurred while processing video {video_url}: {e}")

# Save transcription data and generate audio chunks
def save_transcription_and_chunks(video_id, audio_path, transcription):
    conn = sqlite3.connect(DB_PATH)
    cursor = conn.cursor()

    audio = AudioSegment.from_file(audio_path)

    for entry in transcription:
        start_time = entry['start'] * 1000
        end_time = entry.get('end', start_time + 1000) * 1000
        text = entry['text']

        # Extract audio chunk
        audio_segment = audio[start_time:end_time]
        chunk_filename = f"chunk_{video_id}_{int(start_time)}_{int(end_time)}.wav"
        chunk_path = os.path.join(OUTPUT_DIR, chunk_filename)
        audio_segment.export(chunk_path, format="wav")

        # Insert into transcriptions table
        cursor.execute("INSERT INTO transcriptions (video_id, start_time, end_time, text, audio_chunk_path) VALUES (?, ?, ?, ?, ?)",
                       (video_id, start_time / 1000, end_time / 1000, text, chunk_path))

    conn.commit()
    conn.close()

# Process videos concurrently
def process_videos_concurrently(video_list):
    with ThreadPoolExecutor(max_workers=CHUNK_SIZE) as executor:
        futures = [executor.submit(process_video, video_id, video_url) for video_id, video_url in video_list]
        for future in futures:
            future.result()

# Main function
def main():
    setup_database()

    # Read URLs from text file and insert them into the database
    with open('video_urls.txt', 'r') as f:
        video_urls = [line.strip() for line in f.readlines()]
    insert_video_urls(video_urls)

    # Fetch unprocessed videos
    unprocessed_videos = fetch_unprocessed_videos()

    # Process videos with threading
    process_videos_concurrently(unprocessed_videos)

if __name__ == "__main__":
    main()
```
